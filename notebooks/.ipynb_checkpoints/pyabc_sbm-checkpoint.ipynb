```python
def model(parameter):
    sh = 3
    rates = dict()
    rates["PhiG"]=10**(parameter["PhiG"]-sh)
    rates["PhiGStar"]=10**(parameter["PhiG"]-sh)
    rates["PhiM"]=10**(parameter["PhiM"]-sh)
    rates["KappaG"]=10**(parameter["KappaG"]-sh)
    rates["KappaM"]=10**(parameter["KappaM"]-sh)
    rates["Rho"]=10**(parameter["Rho"]-sh)
    rates["Gamma"]=10**(parameter["Gamma"]-sh)
    rates["KappaExt"]=10**(parameter["KappaExt"]-sh)
        
    weights = [13,13,111,15,15,13]
    
    cells = dict()
    size = 1e+03
    cells['ALIVEEXTERNCANDIDA'] = size
    cells['MONOCYTES'] = 0.5*size
    cells['GRANULOCYTES'] = 5*size
    cells["DECORATED"] = 0
    
    output = ps.run_sbm(1,0.5,481,cells,rates,False)
    sel1 = output[0, [0,2*10,2*20,2*40,2*60,2*90,2*120,2*180], :][:, [0,1,3,4,2]] # C_A, C_K, C_M, C_N, C_E
    sel1 = sel1*weights[:5]
    out_sim1 = np.concatenate([sel1[:,0], sel1[:,1], sel1[:,2], sel1[:,3], sel1[:,4]])

    rates["PhiG"] = 10**(parameter["PhiGD"]-sh)
    
    output2 = ps.run_sbm(1,0.5,481,cells,rates,False)
    E2 = output2[0,2*60,:][[2]][0]
    out_sim2 = E2*weights[5]*np.ones(10)
    
    out_sim = np.concatenate([out_sim1,out_sim2])
    
    return {"data": out_sim}
```

```python
prior = pyabc.Distribution(PhiG=pyabc.RV("uniform", 0, 4),
                           PhiGD=pyabc.RV("uniform", 0, 4),
                           PhiM=pyabc.RV("uniform", 0, 4),
                           KappaM=pyabc.RV("uniform", 0, 4),
                           KappaG=pyabc.RV("uniform", 0, 4),
                           Rho=pyabc.RV("uniform", 0, 4),
                           Gamma=pyabc.RV("uniform", 0, 4),
                           KappaExt=pyabc.RV("uniform", 0, 4))

def distance(x, x0):
    return np.mean((x['data'] - x0['data'])**2)
```

```python
# trial parameters
par = {
    "PhiG": np.log10(0.0321981)+3,
    "PhiGD": np.log10(0.0321981)+3,
    "PhiM": np.log10(0.0137036)+3,
    "KappaM": np.log10(0.501573)+3,
    "KappaG": np.log10(0.0467262)+3,
    "Rho": np.log10(0.00459136)+3,
    "Gamma": np.log10(0.0093675)+3,
    "KappaExt": np.log10(0.176844)+3}
result = model(par)
```

```python
result
```

```python
distance(result,{"data":obs})
```

```python
# Non-adaptive Euclidean distance with uniform weights
#dist_pyABC = pyabc.PNormDistance(p=2)

abc = pyabc.ABCSMC(
    models=model,
    parameter_priors=prior,
    distance_function=distance,
    population_size=2000,
)

db_path = os.path.join(os.getcwd()+'/abc_data/temp.db')
observation = obs
```

```python
db_path
```

```python
df_list = list()
for chain in tqdm(range(2)):
    abc.new("sqlite:///" + db_path, {"data": observation})
    history = abc.run(minimum_epsilon=0.25, max_nr_populations=50)#0.001
    w,df=history.get_distribution(m=0)
    df = w.reset_index().rename_axis("draw")[["Gamma","KappaExt","KappaG","KappaM","PhiG","PhiM","Rho","PhiGD"]].assign(chain=chain).reset_index().set_index(["chain","draw"])
    df_list.append(df)
```

```python
def func_model(df):
    ddd = model({key:val for key,val in list(zip(["Gamma","KappaExt","KappaG","KappaM","PhiG","PhiM","Rho","PhiGD"],df.values))})
    return ddd["data"]
list_xr = list()
num_samples = 500
for j in range(2):
    dat = df_list[j].sample(num_samples)
    list_samples = list()
    list_samples= Parallel(n_jobs=-1)(delayed(func_model)(dat.iloc[i] ) for i in tqdm(range(num_samples)))
    list_xr.append(np.stack(list_samples))
```

```python
fulldf=np.stack(list_xr)
#fulldf.shape
xr_data = xr.DataArray(fulldf, 
        coords={'simulator_dim_3': np.arange(50),"chain":np.arange(2),"draw":np.arange(num_samples)}, 
        dims=['chain','draw','simulator_dim_3'])
xr_data.name = 'simulator'
```

```python
name = 'PYABC_sbm_new_noDecData_2exp_eps0.25_w[13,13,111,15,15,13]_L2'
dataset = az.convert_to_inference_data(xr.Dataset.from_dataframe(pd.concat(df_list)))
dataset.extend(az.convert_to_inference_data(xr_data,group='posterior_predictive'))
dataset.to_netcdf('abc_data/'+name+'.nc')
dataset
```

```python
params_name = list(dataset.posterior.data_vars.keys())
params_name
```

```python
# Parameter values used to simulate the synthetic data
gt_values = []
gt_values.append(0.0193)  # Gamma
gt_values.append(0.2607)   # KappaExt
gt_values.append(0.0469)  # KappaG
gt_values.append(0.0359)   # KappaM
gt_values.append(0.0283)  # PhiG
gt_values.append(0.0125)  # PhiM
gt_values.append(0.0045) # Rho 
gt_values.append(0.0283)  # PhiGD
```

```python
gt_log_values = []
for it,val in enumerate(gt_values):
    gt_log_values.append(np.log10(val))
gt_log_values    
```

```python
start_time = time.time()

fig,axes = plt.subplots(nrows=1,ncols=4,figsize=(12,3))
i = 0
for ax in axes:
    if i<len(params_name):
        par_val = gt_log_values[i]
        par_nam = params_name[i]
        
        df, w = history.get_distribution(m=0, t=history.max_t)
        pyabc.visualization.plot_kde_1d(df-3, w, xmin=-3.2, xmax=1.2, x=par_nam, xname=par_nam, ax=ax, numx=200)#, label=f"PDF t={t}")
                
        ax.axvline(par_val, color="k", linestyle="dashed")
        #ax.legend()
    
    i += 1
    
#filename = os.path.join('figures/', 'post_sbm_IEt_wlog_eps1_GammaR_1e3size_1.svg')
#plt.savefig(filename, format='svg',bbox_inches='tight')    
    
fig,axes = plt.subplots(nrows=1,ncols=4,figsize=(12,3))
for ax in axes:
    if i<len(params_name):
        
        par_nam = params_name[i]
        
        df, w = history.get_distribution(m=0, t=history.max_t)
        pyabc.visualization.plot_kde_1d(df-3, w, xmin=-3.2, xmax=1.2, x=par_nam, xname=par_nam, ax=ax, numx=200)#, label=f"PDF t={t+1}")

        if i<len(gt_log_values):
            par_val = gt_log_values[i]
            ax.axvline(par_val, color="k", linestyle="dashed")
    
    i += 1    
    
#ax.legend(loc='upper right', bbox_to_anchor=(1.5, 1.0))

# Calculate the elapsed time
elapsed_time = time.time() - start_time

#filename = os.path.join('figures/', 'post_sbm_IEt_wlog_eps1_GammaR_1e3size_2.svg')
#plt.savefig(filename, format='svg',bbox_inches='tight')

print(f"The function took {elapsed_time} seconds to execute.") 
```
